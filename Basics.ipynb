```python
p = autos['price'].str.replace('$','')
p = p.str.replace(',','')
p = p.astype(float)
autos['price'] = p
```

```python
o = autos['odometer']
o = o.str.replace(r'[^0-9]+', '')
o = o.astype(int)
autos['odometer'] = o
```

```python
unique_brand_counts = autos["brand"].value_counts()
unique_brand_counts
```

In order to produce comparable mean values for each brand's prices we only select brands with more than 400 occurences in the dataset.

```python
brands = unique_brand_counts[unique_brand_counts > 400].index
mean_prices_brands = {}
for brand in brands:
    brand_mean = autos.loc[autos['brand'] == brand, 'price'].mean()
    mean_prices_brands[brand] = brand_mean
    
sorted(mean_prices_brands.items(), key=lambda item: item[1], reverse=True)
```

cars in the category **sonstige_autos** have a mean price of nearly twice that of the brand **Mini** which ranked second highest in our selected brands data. Coming up are the well-known German brands **Audi**, **Mercedes-Benz** and **BMW** with comparable mean prices ranging from 8900\$ to 8200\$ respectively. **Volkswagen** falls well below this range with a mean price of 5300\$ in this dataset. 


```python
top_brands = unique_brand_counts.head(6).index
mileage_brands = {}
price_brands = {}

for brand in top_brands:
    mileage_mean = autos.loc[autos['brand'] == brand, 'odometer_km'].mean()
    price_mean = autos.loc[autos['brand'] == brand, 'price'].mean()
    mileage_brands[brand] = mileage_mean
    price_brands[brand] = price_mean

mean_milage_brands = pd.Series(mileage_brands).astype(int)
mean_price_brands = pd.Series(price_brands).astype(int)

milage_price_df = pd.DataFrame(mean_milage_brands, columns=['mean_mileage'])
milage_price_df['mean_price'] = mean_price_brands

milage_price_df
    
```

Of the top 6 most frequently sold cars, four are German brands. All six brand's mean_mileage lie close to 130000 km, although mean prices differ widely.
